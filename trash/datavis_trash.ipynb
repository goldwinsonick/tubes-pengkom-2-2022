```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Importing csv data
df = pd.read_csv("daily_weather_data.csv")

# Converting "date" column type to datetime
df["date"] = pd.to_datetime(df["date"], format="%d-%m-%Y")
df2 = df.set_index("date")

print(df.index)
print(df2.index)
```

```python
df3 = df2.loc[df2["country"]=="Indonesia"]["2018-1-1":"2022-1-1"]
test1 = df3.resample("15D")["tavg"].mean()
# print(test1)
df3["tavg"].plot(kind="line",figsize=(15,5))
test1.plot(lw=2)

# .plot(kind="line", figsize=(15,5))
```
